```python
master = transaction.copy()

# Merge user
master = master.merge(user, on='UserId', how='left')

# Merge mode
master = master.merge(mode, on='VisitModeId', how='left')

# Merge attraction
master = master.merge(updated_item, on='AttractionId', how='left')

# Merge attraction type
master = master.merge(type_df, on='AttractionTypeId', how='left')

# Merge City (ONLY CityName)
master = master.merge(
    city[['CityId', 'CityName']],
    on='CityId',
    how='left'
)

# Merge Country (ONLY Country Name)
master = master.merge(
    country[['CountryId', 'Country']],
    on='CountryId',
    how='left'
)

# Merge Region (ONLY Region Name)
master = master.merge(
    region[['RegionId', 'Region']],
    on='RegionId',
    how='left'
)

# Merge Continent (ONLY Continent Name)
master = master.merge(
    continent[['ContinentId', 'Continent']],
    on='ContinentId',
    how='left'
)
```

```python
master.info()
```

```python
master.shape
```

```python
user_visits = master.groupby('UserId')['TransactionId'].count().reset_index()
user_visits.columns = ['UserId', 'TotalVisits']

master = master.merge(user_visits, on='UserId', how='left')
```

```python
user_avg_rating = master.groupby('UserId')['Rating'].mean().reset_index()
user_avg_rating.columns = ['UserId', 'UserAvgRating']

master = master.merge(user_avg_rating, on='UserId', how='left')
```

```python
attraction_popularity = master.groupby('AttractionId')['TransactionId'].count().reset_index()
attraction_popularity.columns = ['AttractionId', 'AttractionPopularity']

master = master.merge(attraction_popularity, on='AttractionId', how='left')
```

```python
attraction_avg_rating = master.groupby('AttractionId')['Rating'].mean().reset_index()
attraction_avg_rating.columns = ['AttractionId', 'AttractionAvgRating']

master = master.merge(attraction_avg_rating, on='AttractionId', how='left')
```
