```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
master = pd.read_csv(r"C:\Users\kisho\Downloads\Tourism_Experience_Analytics_Project\Data\master_tourism_dataset.csv")

master.head()
```

```python
continent_counts = master['Continent'].value_counts()

plt.figure()
continent_counts.plot(kind='bar')
plt.title("User Distribution by Continent")
plt.xlabel("Continent")
plt.ylabel("Number of Visits")
plt.xticks(rotation=45)
plt.show()
```

```python
visitmode_counts = master['VisitMode'].value_counts()

plt.figure()
visitmode_counts.plot(kind='bar')
plt.title("Distribution of Visit Modes")
plt.xlabel("Visit Mode")
plt.ylabel("Number of Visits")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
master['Rating'].plot(kind='hist', bins=5)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
avg_rating_type = master.groupby('AttractionType')['Rating'].mean()

plt.figure()
avg_rating_type.plot(kind='bar')
plt.title("Average Rating by Attraction Type")
plt.xlabel("Attraction Type")
plt.ylabel("Average Rating")
plt.xticks(rotation=90)
plt.show()
```

```python
top_attractions = master['Attraction'].value_counts().head(10)

plt.figure()
top_attractions.plot(kind='bar')
plt.title("Top 10 Most Visited Attractions")
plt.xlabel("Attraction")
plt.ylabel("Visit Count")
plt.xticks(rotation=75)
plt.show()
```

```python
visit_continent = pd.crosstab(master['Continent'], master['VisitMode'])

visit_continent.plot(kind='bar')
plt.title("Visit Mode by Continent")
plt.xlabel("Continent")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_trend = master.groupby('VisitMonth')['TransactionId'].count()

plt.figure()
monthly_trend.plot(kind='bar')
plt.title("Monthly Visit Trend")
plt.xlabel("Month")
plt.ylabel("Number of Visits")
plt.show()
```

```python
yearly_trend = master.groupby('VisitYear')['TransactionId'].count()

plt.figure()
yearly_trend.plot(kind='bar')
plt.title("Yearly Visit Trend")
plt.xlabel("Year")
plt.ylabel("Number of Visits")
plt.show()
```

```python
import seaborn as sns

plt.figure()
sns.heatmap(master[['UserAvgRating','AttractionAvgRating','AttractionPopularity','TotalVisits']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python

visit_type = pd.crosstab(master['AttractionType'], master['VisitMode'])

visit_type.plot(kind='bar', stacked=True)
plt.title("Attraction Type vs Visit Mode")
plt.xticks(rotation=90)
plt.show()
```
